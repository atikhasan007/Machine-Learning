```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(
    n_samples = 100, #row number
    n_features=2,#columns
    n_informative=1,#class prediction 1's column
    n_redundant=0, 
    n_classes=2,#two class 0 and 1
    n_clusters_per_class=1, #1 cluster sorol bivajon
    random_state=41,
    hypercube=False,
    class_sep=20
    )
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```

| Solver        | কাজ করে কীভাবে                               | বাস্তব প্রেক্ষাপটে কবে ব্যবহার করব |
| ------------- | -------------------------------------------- | ---------------------------------- |
| `'liblinear'` | ছোট ডেটার জন্য binary classification         | যখন ফিচার কম                       |
| `'lbfgs'`     | efficient optimization, multivariate support | যখন ডেটা একটু বড়                   |
| `'sag'`       | stochastic, বড় ডেটার জন্য দ্রুত              | বড় ডেটাসেটে                        |
| `'saga'`      | sparse + large dataset, supports all penalty | অনেক ফিচার ও বড় ডেটা               |
| `'newton-cg'` | mathematical derivative ব্যবহার করে          | যখন gradient দরকার হয়              |


| তুমি যদি চাও...        | তাহলে এইটা করো                       |
| ---------------------- | ------------------------------------ |
| Feature কমাও           | `penalty='l1'`, `solver='liblinear'` |
| Weight small রাখো      | `penalty='l2'`, `solver='lbfgs'`     |
| Big dataset handle করো | `solver='saga'`, `sag'`              |
| কোনো penalty ছাড়াই     | `penalty=None`, `solver='lbfgs'`     |


```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty='l2',solver='sag')
lor.fit(X,y)

```

```python
print(lor.coef_)
```

```python
print(lor.intercept_)
```

```python
m1 = lor.coef_[0][0]
m1
```

```python
m2 = lor.coef_[0][1]
m2
```

# Decision boundary

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input+b1
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def gradient_descent(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    gradient_descent
    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
        
    return weights[1:],weights[0]
```

```python
coef_,intercept_ = gradient_descent(X,y)
```

```python
coef_,intercept_
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
